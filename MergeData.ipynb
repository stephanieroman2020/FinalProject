```python
#Setting up dependencies
import pandas as pd
import io
import requests
from datetime import timedelta
```

```python
#Extracting 2018 SPX intraday data
url="https://raw.githubusercontent.com/FutureSharks/financial-data/master/pyfinancialdata/data/stocks/histdata/SPXUSD/DAT_ASCII_SPXUSD_M1_2018.csv"
Data_2018=pd.read_csv(url, sep = ";")
Data_2018.columns = ["Date", "Open", "High", "Low", "Close", "Price"]
Data_2018.head()
```

```python
Data_2018.info()
```

```python
#Extracting 2017 SPX intraday data
url="https://raw.githubusercontent.com/FutureSharks/financial-data/master/pyfinancialdata/data/stocks/histdata/SPXUSD/DAT_ASCII_SPXUSD_M1_2017.csv"
Data_2017=pd.read_csv(url, sep = ";")
Data_2017.columns = ["Date", "Open", "High", "Low", "Close", "Price"]
Data_2017.head()
```

```python
Data_2017.info()
```

```python
#Concatenate 2017 and 2018 data together
SPX_df = pd.concat([Data_2017, Data_2018], ignore_index=True)
SPX_df.head()
```

```python
SPX_df.info()
```

```python
#Extracting only the data we'll need to merge with the Trump Tweets df
SPX_df = SPX_df.drop(columns=['Open', 'High', 'Low', 'Price'])
SPX_df.head(10)
```

```python
#SPX_df['Date'] = SPX_df['Date'].dt.strftime('%Y%m%d %H%M')
#format="%YYYY%mm%dd%H%M%S"
SPX_df2 = pd.to_datetime(SPX_df['Date'])
SPX_df2.head()
```

```python
SPX_df['Date'] = pd.to_datetime(SPX_df['Date'])
SPX_df.head()
```

```python
SPX_df['Date'] = SPX_df['Date'].dt.strftime('%Y%m%d %H%M')
SPX_df.head()
```

```python
SPX_df.info()
```

```python
SPX_df.to_csv('Cleaned_SPX_Data.csv')
SPX_df
```

```python
df = pd.read_csv("mergedata.csv")
df1 = df.drop(columns=['Unnamed: 0'])
df1 = df1.rename(columns={"Date": "Date Tf", "Close": "SPX Tf"}) 

df1.head()
```

```python
# df1[f'Date Plus {Time_Diff} Mins'] = df1['Date'] + timedelta(seconds=60 * Time_Diff)
# # df1[f'Date Plus {Time_Diff} Mins'] = df1[f'Date Plus {Time_Diff} Mins'].dt.strftime('%Y%m%d %H%M')
# SPX_df.head()
# df1.head()

df1['First Date'] = pd.to_datetime(df1['First Date'])
df1['First Date'] = df1['First Date'].dt.strftime('%Y%m%d %H%M')
df2 = df1.rename(columns={"First Date": "Date Ti"}) 
```

```python
df2.head()
```

```python
second_mergedata = SPX_df.merge(df2, left_on='Date', right_on='Date Ti')
second_mergedata = second_mergedata.drop(columns=['Date'])
second_mergedata = second_mergedata.rename(columns={"Close": "SPX Ti"}) 
second_mergedata.head()
```

```python
second_mergedata['% Change SPX'] = ((second_mergedata['SPX Tf'] - second_mergedata['SPX Ti'] ) / second_mergedata['SPX Ti'])*100
```

```python
second_mergedata.head()
```

```python
second_mergedata.sort_values('% Change SPX', ascending=False)
```
